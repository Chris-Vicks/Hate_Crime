```python
import pandas as pd
from scipy.stats import chi2_contingency

# Load your dataset into a DataFrame (replace "data.csv" with your file)
data = pd.read_csv("../data/Hate_crime1\Hate Crimes in TX 1991-2020.csv")

# Select relevant columns
columns_of_interest = ["OFFENSE_NAME", "OFFENDER_RACE", "BIAS_DESC", "LOCATION_NAME"]
subset_data = data[columns_of_interest]

# Create contingency tables for each pair of categorical variables
contingency_tables = {}
for column in subset_data.columns[1:]:
    contingency_table = pd.crosstab(subset_data[column], subset_data["OFFENSE_NAME"])
    contingency_tables[column] = contingency_table

# Calculate chi-squared statistics and p-values
results = {}
for column, table in contingency_tables.items():
    chi2, p, _, _ = chi2_contingency(table)
    results[column] = {"chi2": chi2, "p-value": p}

# Print the results
for column, result in results.items():
    print(f"{column}:")
    print(f"Chi-squared: {result['chi2']}")
    print(f"P-value: {result['p-value']}")
    print()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset into a DataFrame 
data = pd.read_csv("../data/Hate_crime1/Hate Crimes in TX 1991-2020.csv")

# Filter necessary columns
filtered_data = data[["DATA_YEAR", "INCIDENT_ID"]]

# Group and aggregate data
grouped_data = filtered_data.groupby("DATA_YEAR").size()

# Create the line chart
plt.figure(figsize=(10, 6))
plt.plot(grouped_data.index, grouped_data.values, marker='o')
plt.title("Hate Crimes Over the Years in Texas")
plt.xlabel("Year")
plt.ylabel("Number of Hate Crimes")
plt.xticks(rotation=45)
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()

```

```python
data = pd.read_csv("../data/Hate_crime1/Hate Crimes in TX 1991-2020.csv")

# Filter necessary columns
filtered_data = data[["OFFENSE_NAME", "OFFENDER_RACE", "BIAS_DESC", "LOCATION_NAME"]]

# Encode categorical variables with numeric labels
label_encoded_data = filtered_data.apply(lambda x: pd.factorize(x)[0])

# Calculate the correlation matrix
correlation_matrix = label_encoded_data.corr()

# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```
